# Bisection method

The bisection method is one of the simplest methods to find a zero-crossing (f(x) = 0) of a continuous scalar function. It starts from two starting points whose function value have opposite signs. These starting points define an interval in which the zero-crossing is guaranteed to lie. This interval is iterative halved by selecting the subinterval in which the function has a sign change.

The main advantage of this method is its robustness, given two starting points with opposite sign it is guaranteed to converge. Its main disadvantage is that it converges very slow. Therefore often combined with faster methods such as Newton's method (Newton-bisection method) or secant method (Dekker's method).

## Example

Consider the function $f(x) = x^3+3x^2+x-5$

```julia
f(x) =  x.^3+3*x.^2+x.-5.;x = LinRange(-3.,2.,1000); 
plotf(x,f);  
```

```julia
z = -(2-1im);
z.^3+3*z.^2+z.-5.
```
